```python
import numpy as np
import random
import matplotlib.pyplot as plt
# Initialisierung
n=10; w=100000; kWerte=range(n); H=[0]*n
# Simulation
for i in range(w):
    for k in range(n):
        Spiel=[random.randint(0,1) for _ in range(n+k)]
        eins=Spiel[n+k-1]==1 and sum(Spiel[:n+k-1])==n-1
        null=Spiel[n+k-1]==0 and sum(Spiel[:n+k-1])==k
        if eins or null: H[k]+=1
# Ausgabe
plt.bar(kWerte,np.array(H)/w,width=0.2,
        edgecolor='black',color='lightgray')
plt.xlabel('$k$',fontsize=14)
plt.ylabel('rel. HÃ¤ufigkeit',fontsize=14)
plt.show()
```
