```python
import random
import matplotlib.pyplot as plt

# Simuliere ein einzelnes Spiel
def simuliere_spiel(n):
    punkte_anne = 0
    punkte_beate = 0
    
    while punkte_anne < n and punkte_beate < n:
        if random.random() < 0.5:  # Anne gewinnt die Runde
            punkte_anne += 1
        else:  # Beate gewinnt die Runde
            punkte_beate += 1
            
    if punkte_anne == n:
        return punkte_beate  # Beate ist Verliererin
    else:
        return punkte_anne  # Anne ist Verliererin

# Simuliere mehrere Spiele und sammle die Ergebnisse
def simuliere_spiele(n, anzahl_spiele):
    ergebnisse = [0] * n  # Liste, die z채hlt, wie oft die Verliererin 0, 1, ..., n-1 Punkte hat

    for _ in range(anzahl_spiele):
        k = simuliere_spiel(n)  # Punkte der Verliererin
        ergebnisse[k] += 1

    relative_haeufigkeiten = [x / anzahl_spiele for x in ergebnisse]
    return relative_haeufigkeiten

# Zeichne das Stabdiagramm
def plot_ergebnisse(n, relative_haeufigkeiten):
    plt.bar(range(n), relative_haeufigkeiten, tick_label=range(n))
    plt.xlabel('Punkte der Verliererin',fontsize=14)
    plt.ylabel('Relative H채ufigkeit',fontsize=14)
    plt.title(f'Relative H채ufigkeiten der Punkte der Verliererin (n = {n})')
    plt.show()

# Parameter
n = 10  # Anzahl der Punkte zum Sieg
anzahl_spiele = 100000  # Anzahl der Simulationen

# Simulation durchf체hren und Ergebnisse plotten
relative_haeufigkeiten = simuliere_spiele(n, anzahl_spiele)
plot_ergebnisse(n, relative_haeufigkeiten)
```
